### **Load Excel File**

```python
df = pd.read_excel(excel_file_path)
```

```python
df.to_csv("online_retail.csv", index = False)
```

```python
# Now reading the data into spark
df = spark.read.csv("online_retail.csv")
df.show()
```

### **Other settings of read csv**

#### **Exclude infer schema**

```python
df = spark.read.csv("online_retail.csv", inferSchema = True, header = True)
df.show()
df.printSchema()
```

#### **Exclude Header=True**

```python
df = spark.read.csv("online_retail.csv", header = True)
df.show()
```

```python
df.printSchema()
```
